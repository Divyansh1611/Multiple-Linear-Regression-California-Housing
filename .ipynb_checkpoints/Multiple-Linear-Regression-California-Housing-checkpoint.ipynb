## Outlier Treatment

```python
def get_bounedValue(dataset, column_name):
    q1 = dataset[column_name].quantile(0.25)
    q3 = dataset[column_name].quantile(0.75)
    
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    return lower_bound, upper_bound
```

```python
dataset['Price'].describe()
```

```python
dataset.shape
```

```python
for i in dataset.columns:
    lower_bound , upper_bound = get_bounedValue(dataset, i)
#     dataset = dataset[(dataset[i] > lower_bound) & (dataset < upper_bound)]
    dataset = dataset[(dataset[i] > lower_bound) & (dataset[i] < upper_bound)]
    
```

```python
dataset.shape
```

```python
### Extreme outliers are remove, we will plot the distplot graph to check the distribution of all columns
```

```python
# dist plot
t = 1 
plt.figure(figsize =(15, 16))

for i in dataset:
    plt.subplot(4,2,t)
    sns.distplot(dataset[i])
    t += 1
    plt.title(f" {i} Skewness is {dataset[i].skew()}")
    plt.tight_layout()
plt.show()
```

```python
# we can see that after removal of extreme outliers many feature skewness is shifted towards the normal dist.
```

## Transformation 

```python
## All data columns are slightly +ve skewed / right skewed, So I am applying the Square Root Transformation.
for i in ['Price', 'Population', 'MedInc']:
    dataset[i] = np.sqrt(dataset[i])
    
dataset
```

```python
# dist plot
t = 1 
plt.figure(figsize =(15, 16))

for i in dataset:
    plt.subplot(4,2,t)
    sns.distplot(dataset[i])
    t += 1
    plt.title(f" {i} Skewness is {dataset[i].skew()}")
    plt.tight_layout()
plt.show()
```

## Train-Test split

```python
dataset.info()
```

```python
y = dataset['Price']
X = dataset.drop(columns=['Price'])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.33)
```

```python
X_train.shape, X_train.shape, y_train.shape, y_test.shape
```

## Standardization -- Encoding

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
X_train_scaled, X_test_scaled
```

## Linear Regression Model

```python
import statsmodels.api as sm
```

```python
model = sm.OLS(y_train, sm.add_constant(X_train_scaled)).fit()
```

```python
model.summary()
```

```python
# Linear Regression Model using sklearn library
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import *
```

```python
model = LinearRegression()
```

```python
model.fit(X_train_scaled, y_train)
```

```python
# check the model accuracy on train data
y_pre_train = model.predict(X_train_scaled)
```

```python
r2_score(y_train, y_pre_train)
```

```python
# same as in the OLS method
```

```python
#Accuracy on Test Data
y_pred_test = model.predict(X_test_scaled)
y_pred_test
```

```python
# R2 on Test data is 55% means model is able to remove 55% of Error
r2_score(y_test, y_pred_test)
```

```python
print('mse :',mean_squared_error(y_test, y_pred_test))
print('mae :',mean_absolute_error(y_test, y_pred_test))
print('rmse :',np.sqrt(mean_squared_error(y_test, y_pred_test)))
```

## Ridge Regression Model
